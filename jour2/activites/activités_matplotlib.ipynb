```python
import matplotlib.pyplot as plt
```

Visualisation des points

```python
ages=[25,30,38,40,45,48]
salaires=[1100.5,1200.85,1500.95,1800.85,1700.58,2000.75]
plt.plot(ages,salaires,'r*')
plt.xlabel('Age')
plt.ylabel('Salaire')
plt.title('Employés')
plt.show()
```

Visualisation d'une courbe

```python
ages=[25,30,38,40,45,48]
salaires=[1100.5,1200.85,1500.95,1800.85,1700.58,2000.75]
plt.plot(ages,salaires,'r-')
plt.xlabel('Age')
plt.ylabel('Salaire')
plt.title('Employés')
plt.show()
```

Visulisation de plusieurs courbes

```python
import pandas as pd
# Données des employés de l'entreprise 1
df1=pd.DataFrame({'age':[25,30,38,40,45,48],
                'salaire':[1100.5,1200.85,1500.95,1800.85,1700.58,2000.75]})
df1
```

```python
# Données des employés de l'entreprise 2
df2=pd.DataFrame({'age':[31,36,49,51,58],
                'salaire':[1400.5,1500.75,1700.5,1750,1900]})
df2
```

```python
plt.plot(df1['age'],df1['salaire'],'r-',label='Entreprise 1')
plt.plot(df2['age'],df2['salaire'],'b-',label='Entreprise 2')
plt.xlabel('Age')
plt.ylabel('Salaire')
plt.title('Employés')
plt.legend()
plt.show()
```

Visualisation des statistiques

- On reprend les données de ventes des produits utilisées avec les activités de pandas.
- On reprend le calcul de statistiques des quantités vendues par cétgorie.
- Puis, on visualise ces statistiques avec matplotlib sous la forme de chart bar.

```python
import pandas as pd
df_main=pd.read_csv('main.csv')
quantite_vendue_par_categorie = df_main.groupby('categorie')['quantite_vendue'].sum().reset_index()
quantite_vendue_par_categorie
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
quantite_vendue_par_categorie.plot(kind='bar', x='categorie', y='quantite_vendue')

# Ajouter des labels et un titre
plt.xlabel('Produit')
plt.ylabel('Quantité Vendue')
plt.title('Quantités Vendues par Produit')

# Afficher le graphique
plt.show()
```
