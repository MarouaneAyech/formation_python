```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<b>Charger les données</b>

```python
df_main = pd.read_excel("data/data_jour4.xlsx", sheet_name='data_etiquetée')
df_main.head()
```

<b>Histogrammes des colonnes continues (histplot)</b>

L'hitogramme est utile pour visualiser la distribution des données

```python
# Distribution des prix (puht)
plt.figure(figsize=(10, 6))
sns.histplot(df_main['puht'], 
             bins=20, 
             kde=True)
plt.title('Distribution des prix (puht)')
plt.show()
```

```python
# Distribution des quantités vendues
plt.figure(figsize=(10, 6))
sns.histplot(df_main['quantite_vendue'], 
             bins=20, 
             kde=True)
plt.title('Distribution des quantités vendues')
plt.show()
```

<b>Visualiser Box Plot (Boîte à moustaches)</b>

Le box plot est utile pour visualiser la distribution des données, les valeurs aberrantes d'une variable continue

```python
sns.boxplot(data=df_main, y='quantite_vendue')
plt.title('Répartition des Quantités Vendues par Catégorie')
plt.show()
```

Le box plot est utile pour visualiser la distribution des données, les valeurs aberrantes des groupes d'une variable continue.

```python
sns.boxplot(data=df_main, x='categorie', y='quantite_vendue')
plt.title('Répartition des Quantités Vendues par Catégorie')
plt.show()
```

<b>Diagramme de dispersion (Scatter Plot) entre variables continues</b>

Il peut servir pour identifier des relations ou des corrélations

Il peut servir pour identifier des clusters/groupes

```python
sns.scatterplot(data=df_main, x='puht', y='quantite_vendue', hue='categorie')
plt.title('Quantité Vendue vs. PUHT par Catégorie')
plt.show()
```

```python
sns.scatterplot(data=df_main, x='puht', y='quantite_vendue', hue='nom')
plt.title('Quantité Vendue vs. PUHT par Marque')
plt.show()
```

<b>Matrice de corrélation entre des variables continues en utilisant Heatmap</b> 

La heatmap peut être utilisée pour visualiser les matrices de corrélation entre plusieurs variables continues.

```python
corr = df_main[['puht', 'quantite_vendue']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Carte de Chaleur de la Corrélation')
plt.show()
```

<b>Count Plot des variables discrètes</b>

Le count plot est utilisé pour visualiser la fréquence des différentes valeurs d'une variable discrète.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compter le nombre d'observations pour chaque catégorie de 'nom'
sns.countplot(data=df_main, x='nom')
plt.title('Nombre de ventes par Produit')
plt.show()

# Compter le nombre d'observations pour chaque catégorie de 'categorie'
sns.countplot(data=df_main, x='categorie')
plt.title('Nombre de ventes par Catégorie')
plt.show()
```
